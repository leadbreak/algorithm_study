```python
"""
programmers weekly coding challenge - 5th
https://programmers.co.kr/learn/courses/30/lessons/84512
"""

from itertools import product

def solution(word):
    
    dictionary = [''.join(i) for i in list(product(['A', 'E', 'I', 'O', 'U', ' '], repeat=5))]
    
    data = [x.replace(" ", "") for x in dictionary]
    data = list(set(data))
    
    answer_list = sorted(data, reverse=False)
    answer = answer_list.index(word)
    return answer
```
