##4. 송아지 찾기

image.png

```
# 1차 시도 : time-limit으로 인한 60점

import sys, time
import numpy as np
from collections import deque
import itertools

start = time.time()

# sys.stdin = open("input.txt", "r")   

            
if __name__ == '__main__':
      s, e = map(int, input().split())

      dis = [10000] * 10000
      dis[s] = 0

      Q = deque()
      Q.append(s)

      while Q :
            now = Q.popleft()

            if now == e :
                  break

            for next in (now+5, now+1, now-1) :
                  if 0 < next <= 10000 :
                        dis[next] = min(dis[now] + 1,dis[next])
                        Q.append(next)

      print(dis[e])      

      # print("총 작업 시간 :", time.time() - start)
```

```
import sys, time
import numpy as np
from collections import deque
import itertools

start = time.time()

# sys.stdin = open("input.txt", "r")   

            
if __name__ == '__main__':

      n = int(input())

      grid = [list(map(int, input().split())) for _ in range(n)]
      
      dQ = deque()
      dQ.append((0,0))

      x = 0
      point = n // 2

      res = 0

      while True :
            if x <= point :
                  sy = point - x
                  ey = point + x

                  res += sum(grid[x][sy:ey+1])

            else :
                  sy = point - (n - x - 1)
                  ey = point + (n - x - 1)

                  res += sum(grid[x][sy:ey+1])


            if x == n-1 :
                  break

            x += 1

          
      print(res)


      # print("총 작업 시간 :", time.time() - start)
 
```

```
import sys, time
import numpy as np
from collections import deque
import itertools

start = time.time()

# sys.stdin = open("input.txt", "r")   

            
if __name__ == '__main__':

      n = int(input())

      grid = [list(map(int, input().split())) for _ in range(n)]
      ch = [[0] * n for _ in range(n) ]

      dx = [0,1,0,-1]
      dy = [1,0,-1,0]


      results = []

      wave = 3

      while True :

            g = grid
            
            c = [[0] * n for _ in range(n) ]

            cnt = 0

            for i in range(n) :
                  for j in range(n) :
          

                        if c[i][j] == 0 and g[i][j] > wave:
                              c[i][j] = 1
                              cnt += 1   
                              dQ = deque()
                              dQ.append((i,j))                           

                              while dQ :
                                    tmp = dQ.popleft()

                                    for k in range(4) :
                                          xx = tmp[0] + dx[k]
                                          yy = tmp[1] + dy[k]

                                          if 0 <= xx < n and 0 <= yy < n and c[xx][yy] == 0  and g[xx][yy] > wave:
                                                c[xx][yy] = 1    
                                                dQ.append((xx,yy))
                              
                            
            
            results.append(cnt)
            
            if wave == 100:
                  print(max(results))
                  # print("총 작업 시간 :", time.time() - start)
                  sys.exit(0)
      
            wave += 1

      

      # print("총 작업 시간 :", time.time() - start)

 
```
