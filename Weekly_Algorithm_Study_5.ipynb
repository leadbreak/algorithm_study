```python
def solution(sizes):
    max_lines = []
    min_lines = []
    for size in sizes :
        max_lines.append(max(size))
        min_lines.append(min(size))

    answer = max(max_lines) * max(min_lines)
    return answer
```

```python
# 1ì°¨

import re

def solution(str1, str2):
    str1 = str1.lower()
    str2 = str2.lower()

    list1 = []
    list2 = []

    for i in range(len(str1)-1) :
        target1 = str1[i:i+2]
        if len(target1) == len(re.sub(r"[^a-z]","", target1)) :
            list1.append(str1[i:i+2])

    for i in range(len(str2)-1) :
        target2 = str2[i:i+2]
        if len(target2) == len(re.sub(r"[^a-z]","", target2)) :
            list2.append(str2[i:i+2])
    
    if list1 == list2 :
        return 65536
    else :
        total = 0
        inter = 0
        if len(list1) > len(list2) :
            longer = list1.copy()
            shorter = list2.copy()
        else :
            longer = list2.copy()
            shorter = list1.copy()

        while longer :
            total += 1
            if longer[0] in shorter :

                shorter.remove(longer[0])
                longer.remove(longer[0])                
                inter += 1
            else :
                longer.remove(longer[0])

        total += len(shorter)

        return int((inter/total) * 65536)
```
